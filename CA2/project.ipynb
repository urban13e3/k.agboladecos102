```python
import tkinter as tk
from tkinter import messagebox

menu = {
    'rice/pasta': {
        'Jollof Rice': 350,
        'Coconut Fried Rice': 350,
        'Jollof Spaghetti': 350
    },
    'proteins': {
        'Sweet Chilli Chicken': 1100,
        'Grilled Chicken Wings': 400,
        'Fried Beef': 400,
        'Fried Fish': 500,
        'Boiled Egg': 200,
        'Sauteed Sausages': 200
    },
    'side_dishes': {
        'Savoury Beans': 350,
        'Roasted Sweet Potatoes': 300,
        'Fried Plantains': 150,
        'Mixed Vegetable Salad': 150,
        'Boiled Yam': 150,
    },
    'soups_and_swallows': {
        'Eba': 100,
        'Pounded Yam': 100,
        'Semo': 100,
        'Atama Soup': 450,
        'Egusi Soup': 480
    },
    'beverages': {
        'Water': 200,
        'Glass Drink (35cl)': 150,
        'PET Drink (35cl)': 350,
        'PET Drink(50cl)': 350,
        'Glass/Canned Malt': 500,
        'Fresh Yo': 600,
        'Pineapple Juice': 350,
        'Mango Juice': 350,
        'Zobo Drink': 350
    }
}

def calculate_total():
    total_order = sum(menu_item_price[item] * quantity.get() for item, quantity in order_vars.items())
    discount = 0
    if total_order < 1000:
        discount = 0
    elif total_order < 2500:
        discount = 0.1
    elif total_order < 5000:
        discount = 0.15
    else:
        discount = 0.25
    total_charge = total_order * (1 - discount)
    total_charge_str = f"Total Charges: ₦{int(total_charge)}"
    total_label.config(text=total_charge_str)

root = tk.Tk()
root.title("PAU Cafeteria")

order_vars = {}
menu_item_price = {}


for category, items in menu.items():
    tk.Label(root, text=category.capitalize(), font=("Helvetica", 14, "bold")).pack()
    for item, price in items.items():
        frame = tk.Frame(root)
        frame.pack(anchor="w")
        tk.Label(frame, text=f"{item} - ₦{price}", width=25).pack(side="left")
        order_vars[item] = tk.IntVar()
        tk.Entry(frame, textvariable=order_vars[item], width=5).pack(side="left")
        menu_item_price[item] = price


total_label = tk.Label(root, text="", font=("Helvetica", 14, "bold"))
total_label.pack()

tk.Button(root, text="Submit Order", command=calculate_total, bg="green", fg="white").pack()

root.mainloop()
```
